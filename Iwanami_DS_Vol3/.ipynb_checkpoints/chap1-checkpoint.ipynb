```python
import numpy as np
import statsmodels.api as sm
import statsmodels.tools as tools
import matplotlib.pyplot as plt
```

# 層別解析の例
1. 相関係数0.8くらいの変数x, zを発生させる
2. yを以下のモデルから生成. eは正規乱数
$$
    y = 1.5x + 1.1z + e
$$

```python
np.random.seed(123)

size = 250

x = np.random.randn(size)
z = x + np.random.normal(0, 0.75, size)

print("corr(x, z): {}".format(np.corrcoef(x, z)[0, 1]))

y = 1.5*x + 1.1*z + np.random.randn(size)

plt.scatter(x, y)
plt.show()
```

すべてのデータを用いてまずは以下のモデルでパラメータを推定してみる。

$$
    \hat{y} = \beta_0 + \beta_1 x
$$

```python
model = sm.OLS(y, tools.add_constant(x))
result = model.fit()

print(result.summary())

x_ = np.arange(-3, 3, 0.1)

plt.scatter(x, y)
plt.plot(x_, result.predict(tools.add_constant(x_)), c="r", label="invalid coef")
plt.plot(x_, 1.5*x_, c="b", label="true coef")
plt.legend()
plt.show()
```

傾きが2.67で明らかに課題推定している…  
zの標準偏差で層別（4層）して層別にモデルを推定してみる  

$$
    \hat{y_i} = \beta_{0i} + \beta_{1i} x_i
$$

```python
stratified_index = []

for i in z:
    if i > 0.75:
        stratified_index.append(0)
    elif i > 0:
        stratified_index.append(1)
    elif i > -0.75:
        stratified_index.append(2)
    else:
        stratified_index.append(3)
        
stratified_index = np.array(stratified_index)
        
plt.scatter(x, y, c=stratified_index)
plt.plot(x_, 1.5*x_, c="b")

# 係数を保存
coefs = []

# 推定値を統合するときの重み
# 標準誤差^2の逆数??
weights = []

for i in range(4):
    index_ = np.where(stratified_index==i)
    
    model_ = sm.OLS(y[index_], tools.add_constant(x[index_]))
    result_ = model_.fit()
    coef_ = result_.params[1]
    
    coefs.append(coef_)
    weights.append(1 / result_.bse[1]**2)
    
    print("Group{}\'s coef: {}".format(i+1, coef_))
    
    plt.plot(x_, result_.predict(tools.add_constant(x_)), c="r")

coefs = np.array(coefs)
weights = np.array(weights)
    
plt.show()
```

層別することでさっきよりは真の値に近いものが得られている。  
これらを統合することで、yに対するxの因果効果の推定値が求まる。  
具体的には、回帰係数をその分散の逆数で重みをつけた平均をとる。

```python
print("coef: {}".format(np.average(coefs, weights=weights)))
```
